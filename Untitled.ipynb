```python
taser_incidents_url = "https://docs.google.com/spreadsheets/d/1KXS61zifZU_8LgRK3bLUvQI3XFkz_k7n_gkkBNGJBJo/pub?gid=454033036&single=true&output=csv"
taser_incidents = pd.read_csv(taser_incidents_url)
```

```python
per_capita = taser_incidents.merge(population, left_on="department", right_on="town")[["department","town","incidents","population"]]
```

```python
per_capita["incidents_per_c"] = per_capita.apply(lambda x: float(x["incidents"]) * 10000.0 / float(x["population"]), axis=1).fillna("N/A")
```

```python
per_capita[["department","incidents","incidents_per_c"]].to_csv("per_capita.tsv",sep="\t")
```
